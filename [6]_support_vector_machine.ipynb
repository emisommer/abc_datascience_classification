# Support Vector Machine (SVM)

image.png

SVMs are powerful supervised machine learning algorithms used primarily for classification tasks.  Let's understand their core concepts:

* **Finding the Best Boundary**: An SVM aims to find the optimal hyperplane (a decision boundary) that cleanly separates data points belonging to different classes.

* **Maximizing Margins**: SVMs don't just find any separating hyperplane; they seek to maximize the margin. The margin is the distance between the hyperplane and the nearest data points from each class (called support vectors).  A larger margin generally results in a more robust model, less prone to overfitting.

* **Handling Non-Linearity with Kernels**:  What if the data isn't linearly separable?  SVMs use a clever trick called kernels.  Kernels (like 'rbf', 'poly')  transform data into a higher-dimensional space where linear separation may become possible.

### Key Advantages

* **Robustness**: Focus on support vectors and maximizing margins helps SVMs generalize well, even in the presence of some noise.

* **Non-linearity**: Kernels elegantly handle non-linearly separable datasets.

* **Versatility**: SVMs can be applied to classification as well as regression problems (Support Vector Regression - SVR).


### Common Applications

* Image Classification
* Text Categorization
* Bioinformatics
* Time Series Forecasting

```
# Support Vector Machine (SVM)

## Importing the libraries

import numpy as np  # For numerical calculations
import matplotlib.pyplot as plt  # For data visualization
import pandas as pd  # For data handling

## Importing the dataset

dataset = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ-GzTqfan25hVJ4BrsZr0dimNorFVk7OQ49yGl39SWEEtL8pZ6qlSzZczV2jhg32Zni8AiPw0B3e-W/pub?gid=1606731410&single=true&output=csv')
X = dataset.iloc[:, :-1].values  # Extract feature variables
y = dataset.iloc[:, -1].values  # Extract the target variable (class labels)

## Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
# Splits the data into 75% for training and 25% for testing (random_state ensures consistency)

## Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # Create a scaler object
X_train = sc.fit_transform(X_train)  # Standardize the training set
X_test = sc.transform(X_test)  # Standardize the testing set using the same scaler

## Training the SVM model on the Training set

from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state=0)  # Create an SVM classifier with a linear kernel
classifier.fit(X_train, y_train)  # Train the SVM model

## Making the Confusion Matrix

from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)  # Generate predictions on the test set
cm = confusion_matrix(y_test, y_pred)  # Calculate the confusion matrix
print(cm)
accuracy_score(y_test, y_pred)  # Calculate the model's accuracy
```

## Key Points:

* **SVM Goal**: Support Vector Machines aim to find the optimal hyperplane that separates classes in the feature space, while maximizing the margin (distance) between the hyperplane and the nearest data points (support vectors).

* **Linear Kernel**:  In this code, `kernel='linear'` specifies a linear SVM. This is suitable when the data is linearly separable.

* **Non-linear Problems**:  For non-linearly separable data, consider kernels like:

`'rbf'` (Radial Basis Function)
`'poly'` (Polynomial)
